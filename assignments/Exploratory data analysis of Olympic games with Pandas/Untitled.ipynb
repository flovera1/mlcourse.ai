```python
import pandas as pd
```

```python
PATH = './data/athlete_events.csv'
```

```python
data = pd.read_csv(PATH)
data.head()
```

```python
# 1. How old were the youngest male and female participants of the 1996 Olympics?
```

```python
ndata = data[data['Year'] == 1996]
```

```python
ndata[ndata['Sex'] == "M"].sort_values(by='Age', ascending=True).head() # this shows the man would be 14 years old
```

```python
ndata[ndata['Sex'] == "F"].sort_values(by='Age', ascending=True).head() # this shows the man would be 12 years old
```
